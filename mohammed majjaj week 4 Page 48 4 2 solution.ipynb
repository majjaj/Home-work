Exercise: The numbers generated by numpy.random.random are supposed to be uniform between 0 and 1; that is, every value in the range should have the same probability.

Generate 1000 numbers from numpy.random.random 

```python
import numpy as np
import random

from matplotlib import pyplot as plt
%matplotlib inline
import thinkstats2
import thinkplot

r = np.random.random(1000)
```

```python
pmf = thinkstats2.Pmf(r, label='ran numbr')
```

plot their PMF.

```python
thinkplot.hist(pmf)
thinkplot.Config(xlabel='ran numbr', ylabel='PMF')
```

What goes wrong?

the graph shows no values ???

```python
import pandas as pd
```

```python
df = pd.DataFrame(pmf.Items(), columns = ['randm_numbr', 'probability'])
```

```python
df.head()
```

```python
df.count()
```

```python
df.nunique()
```

we can see that all random numbers generated by np.random.random are unique. so each number appears one time.
each number will have the same probability to happen wich is 1/1000  = 0.001

```python
plt.bar(df['randm_numbr'], df['probability'],  
        width = 0.0008) 
plt.xlabel("randm_numbr") 
plt.ylabel("PMF") 
```

the above is my solution to vizualise that all random numbers had the same probability 0.001

Now plot the CDF. Is the distribution uniform?

answer is yes.

```python
thinkplot.Cdf(thinkstats2.Cdf(r, label='cdf'))
thinkplot.Config(xlabel='Birth weight (pounds)', ylabel='CDF')
```
