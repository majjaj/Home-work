```python
import sys
import numpy as np
import thinkstats2
from collections import defaultdict
dct_file='D:/data science/classes/DSC530 Exploratory_Data_Analysis/thinkstats/ThinkStats2-master/code/2002FemResp.dct'
dat_file='D:/data science/classes/DSC530 Exploratory_Data_Analysis/thinkstats/ThinkStats2-master/code/2002FemResp.dat.gz'

dct_file2='D:/data science/classes/DSC530 Exploratory_Data_Analysis/thinkstats/ThinkStats2-master/code/2002FemPreg.dct'
dat_file2='D:/data science/classes/DSC530 Exploratory_Data_Analysis/thinkstats/ThinkStats2-master/code/2002FemPreg.dat.gz'

nrows=None

#df: respondent DataFrame
   
    
dct = thinkstats2.ReadStataDct(dct_file)
df = dct.ReadFixedWidth(dat_file, compression='gzip', nrows=nrows)

#df2: pregnancy DataFrame

dct2 = thinkstats2.ReadStataDct(dct_file2)
df2 = dct2.ReadFixedWidth(dat_file2, compression='gzip', nrows=nrows)
```

pregnum is a record that indicates how many times each respondent has been pregnant. 

will count the values and compare to results published in the [codebook](https://www.icpsr.umich.edu/nsfg6/Controller?displayPage=labelDetails&fileCode=FEM&section=R&subSec=7869&srtLabel=606835)

```python
pregnum_values = df.pregnum.value_counts().sort_index()
print(pregnum_values)
count_7_95 = pregnum_values[7:96].sum()
print('count_7_95 :' + str(count_7_95))
print('total :' + str(pregnum_values.sum()))
```

```python
#Make a map from caseid to list of preg indices.
#df2_map: dict that maps from caseid to list of indices into `df2`

d = defaultdict(list)
for index, caseid in df2.caseid.iteritems():
    d[caseid].append(index)
df2_map = d
```

```python
##Validate pregnum in the respondent file
invalid=0
valid = 0
for index, pregnum in df.pregnum.items():
    caseid = df.caseid[index]
    indices = df2_map[caseid]

        # check that pregnum from the respondent file equals
        # the number of records in the pregnancy file
    if len(indices) != pregnum:
      
        invalid=invalid+1

    else: 
        valid = valid + 1
        
print('invalid count :' + str(invalid))
print('valid count   :' + str(valid))
#return True
```

cross validation above confirm that number of pregnacy by caseid matches between respendent and pregnancy files
